```
import math, random

#Função binário para decimal
def binDec(b):
    d = 0
    for i in range(len(b)):
      d+=int(b[i])*(2**((len(b)-1)-i))
    return d

#Formatação do Endereço
def formatEndereco():
    global endereco
    global endTag
    global endConjunto
    global endCelula
    for i in range(len(endereco)):
        if i >= 0 and i < tagBits:
            endTag = endTag + endereco[i] 
        elif i >= tagBits and i < (tagBits + conjBits):
            endConjunto = endConjunto + endereco[i] 
        elif i >= (tagBits + conjBits):
            endCelula = endCelula + endereco[i] 
    print("\nFormato de endereço da cache:"+
              "\n| Tag      |",endTag,
              "\n| Conjunto |",endConjunto,
              "\n| Célula   |",endCelula)

# Função de busca da Tag
def searchTag():
    global indBusca 
    busca = False
    for indBusca in range((conjunto * LpConj), ((conjunto * LpConj) + LpConj)): 
        print("\nLinha",indBusca,":")
        print("| Tag| Célula 1 | Célula 0 |")
        formatLinha(indBusca) 
        if endTag == cache[indBusca][len(cache[indBusca])-1]: 
            busca = True
            break
    return busca
    
#Função de atualização de linha da cache
def novaCache():
    global endTag
    global endConjunto
    global linha
    bloco = binDec(endTag+endConjunto) 
    print("\nBloco buscado na MP:", endTag+endConjunto)
    linha = random.randrange((conjunto * LpConj), ((conjunto * LpConj) + LpConj)) 
    print("Linha atualizada na Cache:", linha)
    j = 0
    for i in range((bloco * X), ((bloco * X) + X)): 
        cache[linha][j] = mp[i] 
        j += 1
    cache[linha][(len(cache[linha])-1)] = endTag 

#Função de saída dos dados da linha
def formatLinha(l):
    saida = "| "
    for j in range(len(cache[l])-1, -1, -1):
        saida = saida + cache[l][j] + " | "
    print(saida)
    
mp = []       #Memória Principal
cellMp = 16   #Qtd de células da MP
lrgLin = 2    #Largura das linhas da cache
CapCache = 8  #Capacidade da cache
L = 4         #Quantidade de linhas da cache
LpConj = 2    #Quantidade de linhas por conjunto

B = int(cellMp / lrgLin)             # B = 8
L = int(CapCache / lrgLin)           # L = 4
C = int(L / LpConj)                  # C = 2
BlcConj = int(B / C)                 # BlcConj = 4
tagBits = int(math.log2(BlcConj))    # tagBits = 2 bits
conjBits = int(math.log2(C))         # conjBits = 1 bit
celulaBits = int(math.log2(lrgLin))  # celulaBits = 1 bit

#Memória Cache: Célula 0 | Célula 1 | Tag
cache = [["00000000", "00000000", "00"],
         ["00000000", "00000000", "01"],
         ["00000000", "00000000", "00"],
         ["00000000", "00000000", "01"]]

print("\nMemória Cache\n\n| Tag | Célula 1 | Célula 0 |\n")
for i in range(len(cache)): 
    formatLinha(i) 

# Dados da Memória Principal
print("\nMemória Principal\n")
for i in range(cellMp):
  mp.append("00000000") 
  print(mp[i])

#Inputs do usuário
while True:
  comando = input("\nDigite W para escrever, R para ler, L para listar toda a memória e qualquer tecla para parar: ") 

  # ESCREVER
  if comando=="W" or comando=="w":
    endereco = input("Digite o endereço: ") 
    dado = input("Digite o dado: ")
      
    endTag = ""
    endConjunto = ""
    endCelula = ""
    formatEndereco() 
    
    #Escrita na Cache
    conjunto = binDec(endConjunto) 
    print("\nEscrita no conjunto",conjunto,"da Cache:")
      
    if(searchTag() == True): 
        print("Bloco "+endTag+endConjunto+" encontrado na Cache")
        linha = indBusca 
    else:
        print("Bloco "+endTag+endConjunto+" não encontrado na Cache")
        novaCache() 
    
    cache[linha][binDec(endCelula)] = dado 
    print("\nLinha", linha, "<--- " + dado)
    print("| Tag| Célula 1 | Célula 0 |")
    formatLinha(linha) 
    
    #Escrita na MP
    print("\nEscrita na Memória Principal\nMP <--- Cache")
    mp[binDec(endereco)] = cache[linha][binDec(endCelula)] 
    print("Endereço " + endereco + " <--- " + cache[linha][binDec(endCelula)])
    print("")
          

  #Leitura
  elif comando == "R" or comando == "r":
    endereco = input("Digite o endereço: ") 
    
    endTag = ""
    endConjunto = ""
    endCelula = ""
    formatEndereco()   

    #Busca da tag 
    print("\nAcesso ao endereço na Memória Cache para Leitura")
    conjunto = binDec(endConjunto) 
    print("Busca no Conjunto",conjunto)
    
    if searchTag() == True:  
      print("\nHit!")
      linha = indBusca 
      print("\nInformação na Cache:\nEndereço: " + endereco + "\nDado: " + cache[linha][binDec(endCelula)])
      print("")
    else: 
      print("\nMiss!\nO bloco está sendo buscado na Memória Principal")
      novaCache() 
      print("| Tag | Célula 1 | Célula 0 |")
      formatLinha(linha) 
      print("\nEndereço: " + endereco + "\nDado: " + cache[linha][binDec(endCelula)])
            
            
  #Listar
  elif comando == "L" or comando == "l":
    print("\nMemória Cache:\n| Tag | Célula 1 | Célula 0 |\n-----------------------------")
    for i in range(len(cache)): 
      formatLinha(i) 
    print("\nMemória Principal:")
    for i in range(cellMp):
      print(mp[i])
            
  #Fim
  else:
    print("Você saiu do programa")
    break
```
