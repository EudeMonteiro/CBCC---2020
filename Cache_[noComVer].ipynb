```
import math, random

def convBinDec(b):
    d = 0
    for i in range(len(b)):
        d += int(b[i]) * (2 ** ((len(b) - 1) - i))
    return d


def quebraEndereco():
    global endereco
    global endTag
    global endConjunto
    global endCelula
    for i in range(len(endereco)):
        if i >= 0 and i < tagBits:
            endTag = endTag + endereco[i] 
        elif i >= tagBits and i < (tagBits + conjuntoBits):
            endConjunto = endConjunto + endereco[i] 
        elif i >= (tagBits + conjuntoBits):
            endCelula = endCelula + endereco[i] 
    print("\nFormato de endereço da cache com mapeamento associativo por conjunto de",LpConj,":"+
              "\n| Tag      |",endTag,
              "\n| Conjunto |",endConjunto,
              "\n| Célula     |",endCelula)


def buscaTag():
    global iBusca 
    busca = False
    for iBusca in range((conjunto * LpConj), ((conjunto * LpConj) + LpConj)): 
        print("Linha",iBusca,":")
        print("| Tag |  Célula 1  |  Célula 0  |")
        exibeLinha(iBusca) 
        if endTag == cache[iBusca][len(cache[iBusca])-1]: 
            busca = True
            break
    return busca
    

def atualizaCache():
    global endTag
    global endConjunto
    global linha
    bloco = convBinDec(endTag+endConjunto) 
    print("\nBloco buscado na MP:", endTag+endConjunto)
    linha = random.randrange((conjunto * LpConj), ((conjunto * LpConj) + LpConj)) 
    print("Linha atualizada na Cache:", linha)
    j = 0
    for i in range((bloco * X), ((bloco * X) + X)): 
        cache[linha][j] = mp[i] 
        j += 1
    cache[linha][(len(cache[linha])-1)] = endTag 


def exibeLinha(l):
    saida = "| "
    for j in range(len(cache[l])-1, -1, -1):
        saida = saida + cache[l][j] + " | "
    print(saida)
    
mp = [] 
Np = 16  
X = 2  
Ncache = 8  
L = 4   
LpConj = 2  


B = int(Np / X)  

L = int(Ncache / X) 

C = int(L / LpConj)  

BpConj = int(B / C)  

tagBits = int(math.log2(BpConj))  

conjuntoBits = int(math.log2(C)) 

celulaBits = int(math.log2(X))  


cache = []
cache.append(["00000000", "00000000", "000"])  
cache.append(["00000000", "00000000", "001"])  
cache.append(["00000000", "00000000", "000"])  
cache.append(["00000000", "00000000", "001"])  

print("\nMemória Cache\n| Tag |  Célula 1  |  Célula 0  |\n-----------------------------")
for i in range(len(cache)): 
    exibeLinha(i) 


print("\nMemória Principal")
for i in range(Np):
    mp.append("00000000") 
    print(mp[i])


while True:
    comando = input("\nDigite W para escrever, R para ler, L para listar toda a memória e qualquer tecla para parar: ") 

    
    if comando == "W" or comando == "w":
        while True:
            try:
                endereco = input("Digite o endereço de 4 bits: ") 
                
                for i in range(len(endereco)):
                    if endereco[i] != "0" and endereco[i] != "1":
                        raise Exception
                        break
                if len(endereco) != 4: 
                    raise Exception
                break
            except Exception: 
                print("Endereço inválido. Tente novamente.")

        
        while True:
            try:
                dado = input("Digite o dado de 8 bits: ")
                for i in range(len(dado)):
                    if dado[i] != "0" and dado[i] != "1": 
                        raise Exception
                        break
                if len(dado) != 8: 
                    raise Exception
                break
            except Exception: 
                print("Dado inválido. Tente novamente")
        
        
        
        endTag = ""
        endConjunto = ""
        endCelula = ""
        quebraEndereco() 
        
        
        conjunto = convBinDec(endConjunto) 
        print("\nEscrita no conjunto",conjunto,"da Cache:")
         
        if(buscaTag() == True): 
            print("Bloco "+endTag+endConjunto+" encontrado na Cache")
            linha = iBusca 
        else:
            print("Bloco "+endTag+endConjunto+" não encontrado na Cache")
            atualizaCache() 
        
        cache[linha][convBinDec(endCelula)] = dado 
        print("\nLinha", linha, "<--- " + dado)
        print("| Tag |  Célula 1  |  Célula 0  |")
        exibeLinha(linha) 
        
        
        print("\nEscrita na Memória Principal\nMP <--- Cache")
        mp[convBinDec(endereco)] = cache[linha][convBinDec(endCelula)] 
        print("Endereço " + endereco + " <--- " + cache[linha][convBinDec(endCelula)])
        print("")
          

    
    elif comando == "R" or comando == "r":
        while True:
            try:
                endereco = input("Digite o endereço de 4 bits: ") 
                
                for i in range(len(endereco)):
                    if endereco[i] != "0" and endereco[i] != "1":
                        raise Exception
                        break
                if len(endereco) != 4: 
                    raise Exception
                break
            except Exception: 
                print("Endereço inválido!")

        
        endTag = ""
        endConjunto = ""
        endCelula = ""
        quebraEndereco() 

        
        print("\nAcesso ao endereço na Memória Cache para Leitura")
        conjunto = convBinDec(endConjunto) 
        print("Busca no Conjunto",conjunto)
       
        if buscaTag() == True:  
            print("Resultado: Hit")
            linha = iBusca 
            print("\nInformação na Cache:\nEndereço: " + endereco + "\nDado: " + cache[linha][convBinDec(endCelula)])
            print("")
        else: 
            print("Resultado: Miss\nO bloco será buscado na Memória Principal")
            atualizaCache() 
            print("| Tag |  Célula 1  |  Célula 0  |")
            exibeLinha(linha) 
            print("\nEndereço: " + endereco + "\nDado: " + cache[linha][convBinDec(endCelula)])
            
            
    
    elif comando == "L" or comando == "l":
        print("\nMemória Cache:\n| Tag |  Célula 1  |  Célula 0  |\n-----------------------------")
        for i in range(len(cache)): 
            exibeLinha(i) 
        
        print("\nMemória Principal:") 
        for i in range(Np):
            print(mp[i])
            
    
    else:
        print("Você saiu do programa")
        break
```
