```
import math

#Instruções de máquina
 
#0 HLT    0000
#1 LDA Op 0001  
#2 STR Op 0010
#3 ADD Op 0011
#4 Sub Op 0100
#5 JZ  Op 0101
#6 JP  Op 0110
#7 JN  Op 0111
#8 JMP Op 1000
#9 GET Op 1001
#A PRT Op 1010

#Info
mp = []

NumCodOp = 4 #Número de bits do Código de Operação
NumOpr   = 8 #Número de bits do Operando

CodOp = ""
EndOpr = ""

#Funções

def binDec(b):
  d = 0
  for i in range(len(b)):
    d += int(b[i]) * (2 ** ((len(b) - 1) - i))
  return d

def decToBin(d): 
    return bin(d).replace('0b', '') 

#Funções de Instruções

def HLT():
  print("Halt! Programa Encerrado")

def LDA(n): #0001
  global R0
  if R0==[]:
    R0.append(n)
  else:
    R0=[]
    R0.append(n)
  print("Operando transferido para R0 com êxito")
  
def STR(l,n): #0010
  global R0
  l[n]=R0[0]
  print("Valor transferido de R0 para o operando:",R0[0])

def ADD(n): #0011
  global R0
  print("Somando os valores")
  res=binDec(R0[0])+binDec(n)
  R0.pop(0)
  R0.append(decToBin(res))

def Sub(n): #0100
  global R0
  print("Subtraindo os valores")
  res=binDec(R0[0])-binDec(n)
  R0.pop(0)
  R0.append(decToBin(res))

def JZ(n):  #0101
  global R0
  global CI
  if (int(R0[0]))==0:
    CI=n-1
    print('(JZ) O Contador de Instruções foi atualizado para o endereço do operando')
  else:
    print('(JZ) O Jumper nulo não foi atendido. Executando a próxima instrução')
  
def JP(n):  #0110
  global R0
  global CI
  if (int(R0[0]))>0:
    CI=n-1
  print('(JP) O Contador de Instruções foi atualizado para o endereço do operando')

def JN(n):  #0111
  global R0
  global CI
  if (int(R0[0]))<0:
    CI=n-1
    print('(JN) O Contador de Instruções foi atualizado para o endereço do operando')
  else:
    print('O Jumper negativo não foi atendido. Executando a próxima instrução')

def JMP(n): #1000
  global CI
  CI=n-1
  print('(JMP) O Contador de Instruções foi atualizado para o endereço do operando')

def GET(n): #1001
  n=input("Insira o dado ")
  e=input('Insira o endereço onde se deseja armazenar o dado: ')

  mp[binDec(e)]=n
  print('O dado inserido foi armazenado no endereço do operando')


def PRT(n): #1010
  print("Imprimindo o valor:", n)

#Utilitário
def decodificador(c): #c=RI
  global CodOp
  global EndOpr
  for i in range(len(c)):
    if i >= 0 and i < NumCodOp:
      CodOp = CodOp + c[i]
    elif i >= NumCodOp:
      EndOpr = EndOpr + c[i]
  
  if CodOp=='0000':
    HLT()
    return 0

  elif CodOp=='0001':
    LDA(mp[binDec(EndOpr)])
  
  elif CodOp=='0010':
    STR(mp,binDec(EndOpr))
  
  elif CodOp=='0011':
    ADD(mp[binDec(EndOpr)])
  
  elif CodOp=='0100':
    Sub(mp[binDec(EndOpr)])

  elif CodOp=='0101':
    JZ(binDec(EndOpr))

  elif CodOp=='0110':
    JP(binDec(EndOpr))

  elif CodOp=='0111':
    JN(binDec(EndOpr))

  elif CodOp=='1000':
    JMP(binDec(EndOpr))

  elif CodOp=='1001':
    GET(mp[binDec(EndOpr)])
  
  elif CodOp=='1010':
    PRT(mp[binDec(EndOpr)])

def memProg1(): #Mostrar maior de dois números
  global mp
  mp = []
  mp.append('000100001010'), 
  mp.append('010000001011')
  mp.append('011100000101')
  mp.append('101000001010') 
  mp.append('100000000110') 
  mp.append('101000001011') 
  mp.append('000000000000') 
  mp.append('000000000000')
  mp.append('000000000000')
  mp.append('000000000000')
  mp.append('000000000111') 
  mp.append('000000000101') 
  mp.append('000000000000') 
  mp.append('000000000000') 
  mp.append('000000000000')
  mp.append('000000000000')
  print('\nOs dados do Programa 1 foram adicionados com sucesso à MP\n')

def memProg2(): #Mostrar números de 10 a 0
  global mp
  mp = []
  mp.append('101000001011') #0000
  mp.append('000100001011') #0001
  mp.append('010000001010') #0010
  mp.append('001000001011') #0011
  mp.append('101000001011') #0100
  mp.append('010100000111') #0101
  mp.append('100000000010') #0110
  mp.append('000000000000') #0111 #HLT
  mp.append('000000000000') #1000
  mp.append('000000000000') #1001
  mp.append('000000000001') #1010 Dado
  mp.append('000000001010') #1011 Dado
  mp.append('000000000000') #1100
  mp.append('000000000000') #1101
  mp.append('000000000000') #1110
  mp.append('000000000000') #1111
  print('\nOs dados do Programa 2 foram adicionados com sucesso à MP\n')


while True: #Escolha do programa
  k=input("Escolha 1 para carregar os dados do Programa 1,\nescolha 2 para carregar\
 os dados do Programa 2,\ne escolha 3 para opções da memória memória: ")
  
  if k == "1":
    memProg1()
    break

  elif k == "2":
    memProg2()
    break

  elif k == "3":
    mp = []
    for i in range(16):
      mp.append('000000000000')
    print("\nDados da memória definidos para 000000000000 com êxito\n")
    while True:
      a=input("Digite W para escrever, R para ler, L para listar toda a memória e qualquer tecla para parar: ")
      if a == "W" or a == "w":
        while True:
          b=input('Digite o endereço de 4 bits: ')
          if len(list(str(b)))!=4:
            print('O seu endereço deve conter 4 bits obrigatoriamente, tente novamente')
          else: 
            break
        while True:
          newBit=input("Digite o dado de 12 bits: ")
          if len(newBit)!=12:
            print('O seu dado deve conter 12 bits obrigatoriamente, tente novamente')
          else:
            break        
        endr=(int(b,2))
        mp[endr]=newBit;print()
        
      elif a=='R' or a == 'r':
        while True:
          q=(input('Digite o endereço de 4 bits: '))
          if len(list(str(q)))!=4:
            print('O seu endereço deve conter 4 bits obrigatoriamente, tente novamente')
          else:
            print('\n',mp[int(q,2)],'\n',sep='');break
      
      elif a == "L" or a == "l":
        print('\n'+'Todos os estados:'+'\n');print(*mp,sep='\n');print()

      else:
        print('\nVocê saiu do programa.\n')
        break

  else:
    print('\nOpção inválida. Tente novamente')


#CI e R0
CI  = 0
R0  = []
 
#Leitura da memória 
while CI<len(mp):
  if decodificador(mp[CI])==0:
    break
  else:
    decodificador(mp[CI])
    CI+=1
    print("Contador atualizado. CI=",CI,"\n")
    CodOp=""
    EndOpr=""
```
